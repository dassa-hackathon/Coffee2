```python
def get_bold_list(para):
    bold_list = []
    for run in para.runs:
        if run.style.font.bold:
          bold_list.append(run.text)
          continue
    return bold_list
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer
from torch.utils.data import Dataset

# Sample dataset (replace with your own dataset)


# Convert your data to a DataFrame

# Split the data into training and validation sets
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Tokenization and label encoding
tokenizer = AutoTokenizer.from_pretrained("tomaarsen/span-marker-bert-base-uncased-sourcedata")

class NERDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length=512):
        self.data = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data.iloc[idx]["text"]
        labels = self.data.iloc[idx]["labels"]

        # Tokenize the text
        tokens = self.tokenizer(
            text,
            truncation=True,
            padding="max_length",
            max_length=self.max_length,
            return_tensors="pt"
        )

        # Initialize label_ids with zeros
        label_ids = [0] * self.max_length

        for start, end, label in labels:
            start_token = tokens.char_to_token(start)
            end_token = tokens.char_to_token(end - 1)

            if start_token is not None and end_token is not None:
                # Set label_ids for the span
                for i in range(start_token, end_token + 1):
                    label_ids[i] = 1

        return {
            "input_ids": tokens.input_ids[0],
            "attention_mask": tokens.attention_mask[0],
            "labels": label_ids
        }

# Create training and validation datasets
train_dataset = NERDataset(train_df, tokenizer)
val_dataset = NERDataset(val_df, tokenizer)

# Example of accessing a data point
sample_data = train_dataset[0]
print("Input IDs:", sample_data["input_ids"])
print("Attention Mask:", sample_data["attention_mask"])
print("Labels:", sample_data["labels"])
```

```python
from torch.utils.data import DataLoader
batch_size = 32

# Create data loaders
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=batch_size)
```

```python
import torch
from transformers import AutoModelForTokenClassification

# Load your NER model
model = AutoModelForTokenClassification.from_pretrained("bert-base-uncased")

# Initialize your optimizer and loss function (e.g., Adam and CrossEntropyLoss)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)
loss_fn = torch.nn.CrossEntropyLoss()

# Training loop
for epoch in range(10):
    model.train()
    for batch in train_dataloader:
        inputs = torch.tensor(batch["input_ids"])
        attention_mask = torch.tensor(batch["attention_mask"])
        labels = torch.stack(batch["labels"])
        # print (labels)

        # Forward pass
        outputs = model(inputs, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss

        # Backpropagation and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    # Validation loop (optional)
    model.eval()
    with torch.no_grad():
        for batch in val_dataloader:
            inputs = batch["input_ids"]
            attention_mask = batch["attention_mask"]
            labels = torch.stack(batch["labels"])

            # Validation forward pass
            outputs = model(inputs, attention_mask=attention_mask, labels=labels)
            val_loss = outputs.loss

    # Print or log training and validation metrics
    print(f"Epoch {epoch + 1}: Training Loss = {loss:.4f}, Validation Loss = {val_loss:.4f}")

# Save the trained model
model.save_pretrained("path_to_save_model")
```
